```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier,AdaBoostRegressor
from sklearn.model_selection import  train_test_split
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error,mean_squared_error,confusion_matrix,precision_score,recall_score,f1_score
from sklearn.datasets import make_classification,make_regression
from sklearn.tree import  DecisionTreeClassifier,DecisionTreeRegressor
```

```python
x,y=make_classification(n_samples=1000,n_classes=10,n_features=10,n_informative=5,random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
base_model=DecisionTreeClassifier(max_depth=1)
model=AdaBoostClassifier(estimator=base_model,n_estimators=120,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
