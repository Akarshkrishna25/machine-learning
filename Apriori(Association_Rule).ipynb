```python
import pandas as pd
from mlxtend .preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
items_list=[['bread','butter','jam','milk'],
            ["bread","butter","milk"],
            ["bread","juice","curd"],
            ["bread","milk","juice"],
            ["butter","milk","juice"]]
items_list
```

```python
encode=TransactionEncoder()
data=encode.fit(items_list).transform(items_list)
df=pd.DataFrame(data,columns=encode.columns_)
df
```

```python
item_set=apriori(df,min_support=0.5,use_colnames=True)
item_set
```

```python
rules=association_rules(item_set,metric="confidence",min_threshold=0.7)
rules
```

```python
x=rules.iloc[:,[0,1,3,4]]
x
```
