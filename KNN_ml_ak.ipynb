```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,r2_score,mean_absolute_error,mean_squared_error
```

```python
df=pd.read_csv('/content/Iris.csv')
```

```python
df
```

```python
df=df.drop('Id',axis=1)
```

```python
df
```

```python
x=df.drop('Species',axis=1)
x
```

```python
y=df.Species
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred,average='weighted')
```

```python
recall_score(y_test,y_pred,average='macro')
```

```python
f1_score(y_test,y_pred,average='weighted')
```

KNN Regressor



```python
from sklearn.datasets import make_regression,make_classification
x,y=make_regression(n_samples=1000,n_features=10,random_state=42,noise=10)
x1,y1=make_classification(n_samples=1000,n_features=10,random_state=42,n_classes=3,n_informative=3)
```

```python
x1
```

```python
y1
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
model1=KNeighborsRegressor(n_neighbors=5)
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```
