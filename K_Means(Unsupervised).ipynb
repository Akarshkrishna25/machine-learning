```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_iris,load_wine
data=load_iris()
x=data.data
x
```

```python
out=[]
for i in range(1,11):
  model=KMeans(n_clusters=i)
  model.fit(x)
  out.append(model.inertia_)
out
```

```python
pip install kneed
```

```python
from kneed import  KneeLocator
k_values=range(1,11)
kl=KneeLocator(k_values,out,curve='convex',direction='decreasing')
k=kl.elbow
print("Optimized no. of clusters: ",k)
```

```python
plt.plot(k_values,out,marker='o',color='red')
plt.xlabel("No. of clusters")
plt.ylabel("inertia ")
plt.title("Elbow Curve")
plt.show()
```

```python
model=KMeans(n_clusters=k)
model.fit(x)
```

```python
model.inertia_
```

```python
centroid=model.cluster_centers_
```

```python
labels=model.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=labels)
plt.scatter(centroid[:,0],centroid[:,1],marker='x')
plt.xlabel('Sepal_length')
plt.ylabel("Sepal_Width")
plt.title("K_Means_Clustering")
```

```python
plt.scatter(x[:,2],x[:,3],c=labels)
plt.scatter(centroid[:,2],centroid[:,3],marker='x')
plt.xlabel('petal_length')
plt.ylabel("petal_Width")
plt.title("K_Means_Clustering")
```
